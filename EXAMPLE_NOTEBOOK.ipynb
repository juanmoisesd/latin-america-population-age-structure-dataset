## 1. Setup

````python
!pip install pandas matplotlib seaborn numpy -q
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use("seaborn-v0_8-whitegrid")
print("✅ Ready")
````

## 2. Load sample data

Representative sample for demonstration. Download full data from Zenodo.

````python
# Full data at: https://github.com/juanmoisesd/latin-america-population-age-structure-dataset
np.random.seed(42)
countries = ["Brazil","Mexico","Argentina","Colombia","Chile","Peru","Venezuela","Ecuador"]
years = list(range(2015, 2026))
rows = []
for c in countries:
    base = np.random.uniform(8,25)
    for i,y in enumerate(years):
        rows.append({"country":c,"year":y,"value":round(base+i*np.random.uniform(0.1,0.8)+np.random.normal(0,0.5),2),"n_cases":int(np.random.uniform(500,5000))})
df = pd.DataFrame(rows)
print(f"Shape: {df.shape}")
df.head(10)
````

## 3. Exploratory Analysis

````python
print("Statistical Summary:")
df.groupby("country")["value"].agg(["mean","std","min","max"]).round(2)
````

## 4. Visualizations

````python
fig,axes=plt.subplots(1,2,figsize=(14,5))
for c in countries:
    s=df[df["country"]==c]
    axes[0].plot(s["year"],s["value"],marker="o",ms=3,label=c)
axes[0].set_title("Trend by Country (2015-2025)", fontsize=13,fontweight="bold")
axes[0].legend(fontsize=7,ncol=2)
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Value (%)")
pivot=df.pivot_table(values="value",index="country",columns="year",aggfunc="mean").round(1)
sns.heatmap(pivot,ax=axes[1],cmap="YlOrRd",annot=True,fmt=".1f",annot_kws={"size":7},linewidths=0.5)
axes[1].set_title("Heatmap by Country & Year", fontsize=13,fontweight="bold")
plt.tight_layout()
plt.savefig("example_plot.png",dpi=150,bbox_inches="tight")
plt.show()
print("✅ Saved example_plot.png")
````

````python
fig,ax=plt.subplots(figsize=(10,5))
df_last=df[df["year"]==2025].sort_values("value",ascending=False)
bars=ax.barh(df_last["country"],df_last["value"],color=sns.color_palette("RdYlGn_r",len(df_last)))
ax.set_title("Ranking by Country (2025)", fontsize=14,fontweight="bold")
ax.set_xlabel("Value (%)")
for bar,val in zip(bars,df_last["value"]):
    ax.text(bar.get_width()+0.1,bar.get_y()+bar.get_height()/2,f"{val:.1f}%",va="center")
plt.tight_layout()
plt.savefig("ranking_plot.png",dpi=150,bbox_inches="tight")
plt.show()
print("✅ Saved ranking_plot.png")
````
