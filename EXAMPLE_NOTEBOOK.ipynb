````python
!pip install pandas matplotlib seaborn numpy -q
import pandas as pd,matplotlib.pyplot as plt,seaborn as sns,numpy as np
plt.style.use("seaborn-v0_8-whitegrid")
print("Ready")
````

````python
# Full data: https://juanmoisesd.github.io/latin-america-population-age-structure-dataset/
np.random.seed(42)
cc=["Brazil","Mexico","Argentina","Colombia","Chile","Peru"]
rows=[{"country":c,"year":y,"value":round(np.random.uniform(8,25)+i*0.3,2)} for c in cc for i,y in enumerate(range(2015,2026))]
df=pd.DataFrame(rows)
df.head()
````

````python
fig,ax=plt.subplots(figsize=(10,4))
for c in cc:
    s=df[df.country==c]
    ax.plot(s.year,s.value,marker="o",ms=3,label=c)
ax.set_title("Harmonized demographic dataset: population age str",fontweight="bold")
ax.legend(fontsize=8,ncol=3)
plt.tight_layout()
plt.savefig("example_plot.png",dpi=150,bbox_inches="tight")
plt.show()
````
